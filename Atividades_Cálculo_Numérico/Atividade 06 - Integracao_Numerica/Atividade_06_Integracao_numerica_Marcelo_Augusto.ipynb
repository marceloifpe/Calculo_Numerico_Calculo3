```python
import numpy as np
#integral pela regra dos trapézios

# Dados:
#quantidade de sub-intervalos
n = 100

# limites de integração
a = 0
b = 1

# altura de cada trapézio
h = (b - a) / n

# amostra no eixo x:
x = np.linspace(a, b, n + 1)
print(x)
```

```python
#defina a função a ser integrada
def f(r):
    return np.exp(-r**2)
```

```python
integral = 0

for i in range(1, n):
    integral = integral + 2*f(x[i])
integral = (h/2)*(integral + f(x[0]) + f(x[-1]))

print(integral)
```

```python
# usando integração numérica de 1/3 de simpson
integral_s = 0

for i in range(1, n):
    if i%2 == 0:
       integral_s = integral_s + 2*f(x[i])
    else:
       integral_s = integral_s + 4*f(x[i])
integral_s = (h/3)*(integral_s + f(x[0]) + f(x[-1]))

print(integral_s)
```

```python

import matplotlib.pyplot as plt

# Crie um intervalo de valores x para plotagem
x_plot = np.linspace(a, b, 100)

# Calcular os valores y correspondentes usando a função definida
y_plot = f(x_plot)

# Plot da função
plt.plot(x_plot, y_plot)

# Adicionando labels and title
plt.xlabel('r')
plt.ylabel('f(r)')
plt.title('Gráfico da função f(r) = exp(-r^2)')
plt.grid(True)

# Amostrando o gráfico
plt.show()
```
