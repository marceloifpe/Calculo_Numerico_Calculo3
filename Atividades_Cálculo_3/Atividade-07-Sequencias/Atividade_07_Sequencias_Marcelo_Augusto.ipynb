```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define o ponto inicial
p1 = 0.5

# Valores de k usados no teste
k_values = [2.5, 3.3, 3.5, 4]

# Define o número de interações
num_iterations = 100

# Armazena resultado de k no dicionário
results = {}

# Calcula as sequências e mostra gráfico de k
for k in k_values:
    p_values = [p1]
    for _ in range(num_iterations - 1):
        p_next = k * p_values[-1] * (1 - p_values[-1])
        p_values.append(p_next)
    results[k] = p_values

    # Define campo de plotagem
    df = pd.DataFrame({'n': range(1, num_iterations + 1), 'p_n': p_values})

    plt.figure(figsize=(10, 6))
    plt.plot(df['n'], df['p_n'], marker='o', linestyle='-', markersize=2)
    plt.title(f'Sequência para k = {k}')
    plt.xlabel('n')
    plt.ylabel('p_n')
    plt.grid(True)
    plt.ylim(0, 1)
    plt.show()
    plt.close()

print("Gráfico de sequência gerado com sucesso.")
#Explicação: Nessa sequência populacional de insetos é possível observar um comportamento sazonal de repetição, mas com um padrão distinto para cada valor de k:
# k=2.5: A população converge para um único ponto de equilíbrio estável.
# k=3.3: A população entra em um ciclo de período 2, alternando entre dois valores fixos.
# k=3.5: A população exibe um ciclo de período 4, oscilando entre quatro valores distintos.
# k=4: A população apresenta um comportamento caótico, com valores imprevisíveis e sem um padrão repetitivo claro.
# Com isso é possível concluir que mesmo padrões de modelo simples podem gerar variações complexas, que pode iniciar de forma estável até o caos, dependendo do parâmetro k da sequência.
```
